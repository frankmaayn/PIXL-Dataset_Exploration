```python
import pandas as pd 
import seaborn as sb 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('../../Desktop/PIXL_Project/Datasets/Strelley_Pool_left/Strelley_Pool_left_side_Elemental_Map_02_03_2020.csv', header=1)
dataset
```

```python
dataset.describe()
```

```python
# Get the column names

position_names = []
element_names = []
other_names = []
index = 0;
for i in dataset.columns:
    if (index == 0 ):
        position_names.append(i)
        index += 1;
    if (i.find('X') == 1 or i.find('Y') == 1 or i.find('Z') == 1) and (i.find('_%') == -1 and i.find('_err') == -1 and i.find('_int') == -1):
        position_names.append(i)
    if i.find('_%') != -1:
        element_names.append(i)
  
```

```python
position_names
```

```python
element_names
```

```python
# Get the total count for each element
dataset = dataset[element_names]
element_count = []

for i in element_names:
    element_count.append(dataset[i].sum())
element_count
```

```python
plt.bar(x=element_names, height=element_count)
plt.xticks(rotation='vertical')
plt.title('Element Count for Strelley Pool Left Dataset')
plt.xlabel('Element')
plt.ylabel('Count')
plt.figure(figsize=(16,16))
plt.show()
```

```python
#Check the correlation matrix
dataset = dataset[element_names]
corr = dataset.corr()
plt.figure(figsize=(16,16))
sb.heatmap(corr, annot=True)
plt.show()
```
