```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler #Z-score variables
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

```python
#Dataset
dataset = pd.read_csv("Copy of quantified_element_values - Copy of quantified_element_values.csv")
dataset.head()
```

```python
features = ['Mg_%','Al_%','Ca_%','Ti_%','Fe_%','Si_%']

X = dataset[features]
z = StandardScaler()

X[features] = z.fit_transform(X)

EM = GaussianMixture(n_components = 6 )

EM.fit(X)

cluster = EM.predict(X)

cluster_p = EM.predict_proba(X)


#print("SILHOUETTE: ", silhouette_score(X, cluster))

X['cluster'] = cluster

X
```
