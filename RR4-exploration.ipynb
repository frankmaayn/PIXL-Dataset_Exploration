```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

```python
dataset = pd.read_csv('../../Desktop/PIXL_Project/Datasets/RR4/Scotland_map_2.csv',header =1)
dataset
```

```python
dataset.describe()
```

```python
dataset.head()
```

```python
#Extract the wanted columns
element_names = []
position_names = []
other_names = []
unwanted_names = []
for i in dataset.columns:
    if i.find('_%') != -1:
        element_names.append(i)
    elif i == 'X' or i == ' Y' or i == ' Z':
        position_names.append(i)
    elif i.find('_err') == -1 and i.find('_coeff') ==-1 and i.find('_int')==-1 and i.find('_ECF')==-1:
        other_names.append(i)
    else:
        unwanted_names.append(i)
```

```python
element_dataset = dataset[element_names]
element_dataset
```

```python
position_dataset = dataset[position_names]
position_dataset
```

```python
# Sum the columns
element_count = []
for i in element_names:
    element_count.append(dataset[i].sum())
```

```python
#Plot

plt.bar(x=element_names, height=element_count)
plt.xticks(rotation='vertical')
plt.title('Element Count')
plt.xlabel('Element')
plt.ylabel('Count')
plt.figure(figsize=(16,16))
plt.show()
```
