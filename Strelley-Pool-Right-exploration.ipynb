```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sb 
```

```python
#Get dataset
dataset = pd.read_csv('../../Desktop/PIXL_Project/Datasets/Strelley_Pool_right/Strelley_Pool_right_side_Elemental_Map_02_03_2020.csv', header=1)
dataset
```

```python
dataset.describe()
```

```python
element_names = []
position_names = []
other_names = []

for i in dataset.columns:
    if i.find('_%') != -1:
        element_names.append(i)
    elif i.find('_%') == -1 and i.find('_int') == -1 and i.find('_err') ==-1 and (i.find('X') != -1 or i.find('Y') != -1 or i.find('Z') != -1):
        position_names.append(i)
    else:
        other_names.append(i)
```

```python
element_names
```

```python
position_names
```

```python
dataset = dataset[element_names]
dataset
```

```python
element_counts = []

for i in dataset.columns:
    element_counts.append(dataset[i].sum())
    
```

```python
element_counts
```

```python
plt.bar(x=element_names, height=element_counts)
plt.xlabel('Element')
plt.ylabel('Count')
plt.xticks(rotation='vertical')
plt.title('Element Count for Strelley Pool Righ Dataset')
```

```python
#Check the correlation matrix
dataset = dataset[element_names]
corr = dataset.corr()
plt.figure(figsize=(16,16))
sb.heatmap(corr, annot=True)
plt.show()
```
